```julia
function inv_dist(F, x, d)
    Δ = 0
    while true
        if x < F(Δ)
            return Δ
        end
        Δ += d
    end
end

f(x) = x^2

xs = 0:0.01:5
plot(xs, inv_dist.(f, xs, 0.01))
```

## Biegunowa

```julia
Nb(n)=begin
    U,u,R=zeros(n), zeros(n), zeros(n).+2
    for i in 1:n
        while R[i]>1
            a,b=2*rand()-1,2*rand()-1
            R[i]=a^2+b^2
            u[i]=b
            U[i]=a
        end
    end
    X=zeros(2n)
    s=sqrt.(-2*log.(R)./R)
    X[1:n],X[n+1:end] = s.*U, s.*u
    X
end
            
```
